# **The battle of Neighborhoods**

### **Restaurant recommender system in Bangalore**

### **Introduction:**

Bangalore is the IT Hub of the India. Bangalore has the population of over 15-16 million and is one of the largest city in India.

Bangalore is the IT Hub with various IT companies with a diverse in the social and economic status. With the diversity of various people from various regions and countries, it is very essential to have diversity in the regional and international foods. For the variety of the food and products, bangalore can also be the food paradise for the people who love to eat different kinds of foods.


#### **Problem:**

since bangalore is vast variety of food culture, it might be difficult for new comer in bangalore to find the adequate food based on their preferences as they have not any knowledge about that place. The recommender system can be used in combination with app which provide a interface to order or view the food in their surrounding which they might like. Thus recommended system can be useful to save the time and money of the people. In order to create a recommender system some key points need to be addresed and answered:
1. Quantity and variety of food provided in the nerby restraunts.
2. Quality of the food provided.
3. Similarity of food in the restraunts nearby.
4. Cost of the food sharing similarity and quality.


For answering the above question the recommender system is built with problems in minds. The recommender system will cluster the restraunts based on the location, quality and costs of similar products. This will help provide best quality food at minimum cost to the customers.



#### **Audience:**

As it is clear that the target audience for the recommender system are those who are new and dont have that much knowledge of the locality , those whose dont have time to personally visit restraunts and try their each products in order to find the best one and those who rarely used restraunts and wanted to have food from restraunts from particular occassions. Thus the recommender system will help those customer to provide best quality products with low costs. The auidence can be anyone who is exploring the place and want to taste the best qulaity food with variation in taste.
